```python
import pandas as pd
```

```python
data = {
    "Student_ID": [1, 2, 3, 4, 5],
    "Route": ["A", "B", "A", "C", "B"],
    "Pass_Type": ["Monthly", "Daily", "Monthly", "Semester", "Monthly"],
    "Trips_Per_Week": [5, 3, 6, 2, 4]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.describe())
```

```python
pass_count = df["Pass_Type"].value_counts()
print(pass_count)
```

```python
route_count = df["Route"].value_counts()
print(route_count)
```

```python
### Graphical Representation of pass type usage

import matplotlib.pyplot as plt

pass_count.plot(kind="bar")
plt.title("Bus Pass Usage by Pass Type")
plt.xlabel("Pass Type")
plt.ylabel("Number of Students")
plt.show()
```

```python
route_count.plot(kind="bar")
plt.title("Bus Usage by Route")
plt.xlabel("Route")
plt.ylabel("Number of Students")
plt.show()
```
